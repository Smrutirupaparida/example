```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('house.jpg')
rows, cols = image.shape[:2]

# Define transformation matrix (rotation in this case)
angle = 45
M = cv2.getRotationMatrix2D((cols / 2, rows / 2), angle, 1)

# Apply transformation
transformed_image = cv2.warpAffine(image, M, (cols, rows))

# Display original and transformed image
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(122), plt.imshow(cv2.cvtColor(transformed_image, cv2.COLOR_BGR2RGB)), plt.title('Transformed')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('house.jpg')
rows, cols = image.shape[:2]

# Define translation and rotation matrices
M_translate = np.float32([[1, 0, 50], [0, 1, 50]])
M_rotate = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)

# Apply transformations
translated_image = cv2.warpAffine(image, M_translate, (cols, rows))
transformed_image = cv2.warpAffine(translated_image, M_rotate, (cols, rows))

# Display original and transformed image
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(122), plt.imshow(cv2.cvtColor(transformed_image, cv2.COLOR_BGR2RGB)), plt.title('Transformed')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('house.jpg')
rows, cols = image.shape[:2]

# Define scaling and rotation matrices
scale_factor = 1.5
angle = 45
M_scale_rotate = cv2.getRotationMatrix2D((cols / 2, rows / 2), angle, scale_factor)

# Apply transformation
transformed_image = cv2.warpAffine(image, M_scale_rotate, (cols, rows))

# Display original and transformed images
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(122), plt.imshow(cv2.cvtColor(transformed_image, cv2.COLOR_BGR2RGB)), plt.title('Transformed')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('house.jpg')
rows, cols = image.shape[:2]

# Define points for the affine transformation
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2 = np.float32([[10, 100], [200, 50], [100, 250]])

# Get the affine transformation matrix
M = cv2.getAffineTransform(pts1, pts2)

# Apply the affine transformation
transformed_image = cv2.warpAffine(image, M, (cols, rows))

# Display original and transformed images
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(122), plt.imshow(cv2.cvtColor(transformed_image, cv2.COLOR_BGR2RGB)), plt.title('Transformed')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('house.jpg')
rows, cols = image.shape[:2]

# Define points for the projective transformation
pts1 = np.float32([[50, 50], [200, 50], [50, 200], [200, 200]])
pts2 = np.float32([[10, 100], [200, 50], [100, 250], [280, 220]])

# Get the projective transformation matrix
M = cv2.getPerspectiveTransform(pts1, pts2)

# Apply the projective transformation
transformed_image = cv2.warpPerspective(image, M, (cols, rows))

# Display original and transformed images
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(122), plt.imshow(cv2.cvtColor(transformed_image, cv2.COLOR_BGR2RGB)), plt.title('Transformed')
plt.show()
```
